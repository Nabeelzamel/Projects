```python
class Bank:
    def __init__(self):
        self.accounts = {}

    def register_user(self, username, password):
        if username in self.accounts:
            print("Username already exists. Please choose a different one.")
        else:
            self.accounts[username] = {"password": password, "balance": 0}
            print("Registration successful!")

    def login_user(self, username, password):
        if username in self.accounts and self.accounts[username]["password"] == password:
            print("Login successful!")
            return True
        else:
            print("Invalid username or password. Please try again.")
            return False

    def check_balance(self, username):
        if username in self.accounts:
            balance = self.accounts[username]["balance"]
            print(f"Your current balance is: ${balance}")
        else:
            print("User not found. Please log in first.")

    def deposit(self, username, amount):
        if username in self.accounts:
            self.accounts[username]["balance"] += amount
            print(f"Deposit successful! Your new balance is: ${self.accounts[username]['balance']}")
        else:
            print("User not found. Please log in first.")

    def withdraw(self, username, amount):
        if username in self.accounts:
            if self.accounts[username]["balance"] >= amount:
                self.accounts[username]["balance"] -= amount
                print(f"Withdrawal successful! Your new balance is: ${self.accounts[username]['balance']}")
            else:
                print("Insufficient funds.")
        else:
            print("User not found. Please log in first.")


bank_system = Bank()
while True:
        print("\nBank System Menu:")
        print("1. Register")
        print("2. Login")
        print("3. Check Balance")
        print("4. Deposit")
        print("5. Withdraw")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            username = input("Enter your username: ")
            password = input("Enter your password: ")
            bank_system.register_user(username, password)

        elif choice == "2":
            username = input("Enter your username: ")
            password = input("Enter your password: ")
            if bank_system.login_user(username, password):
                while True:
                    print("\nLogged-in Menu:")
                    print("1. Check Balance")
                    print("2. Deposit")
                    print("3. Withdraw")
                    print("4. Logout")

                    user_choice = input("Enter your choice (1-4): ")

                    if user_choice == "1":
                        bank_system.check_balance(username)
                    elif user_choice == "2":
                        amount = float(input("Enter the deposit amount: "))
                        bank_system.deposit(username, amount)
                    elif user_choice == "3":
                        amount = float(input("Enter the withdrawal amount: "))
                        bank_system.withdraw(username, amount)
                    elif user_choice == "4":
                        print("Logged out.")
                        break
                    else:
                        print("Invalid choice. Please enter a number between 1 and 4.")

        elif choice == "3":
            username = input("Enter your username: ")
            bank_system.check_balance(username)

        elif choice == "4":
            username = input("Enter your username: ")
            amount = float(input("Enter the deposit amount: "))
            bank_system.deposit(username, amount)

        elif choice == "5":
            username = input("Enter your username: ")
            amount = float(input("Enter the withdrawal amount: "))
            bank_system.withdraw(username, amount)

        elif choice == "6":
            print("Exiting Bank System. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 6.")

```
